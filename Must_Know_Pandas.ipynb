## It creates an index on the column/columns you are grouping the data on

```python
df.groupby('SEX')[['Y']].sum()
```

```python
df.groupby('SEX')[['Y','S5']].sum()
```

```python
df.groupby('SEX')[['Y','S5']].mean()
```

## Aggregate 2 operations

```python
df.groupby('SEX')[['Y','S5']].agg(['sum','count'])
```

```python
df.groupby(['SEX','S6'])[['Y','S5']].std()
```

## Assign the grouping to a new dataframe

```python
df_grpby=df.groupby(['SEX','S6'])[['Y','S5']].std()
df_grpby
```

```python
df_grpby.reset_index()
```

# Create a new dataframe

```python
data_1=[['Paris',1,2,10,200,300],['New Delhi',10,5872,810,285700,345200],['Rio',4751,245,1570,27,430]]
columns_1=['col_'+str(i) for i in range(1,6)]
columns_1.insert(0,'city')
df_1=pd.DataFrame(data=data_1,columns=columns_1)
df_1
```

```python
data_2=[['Paris',10],['Rio',30],['New Delhi',20]]
df_2=pd.DataFrame(data=data_2,columns=['city','score'])
df_2
```

# Merge, Join and concat

```python
pd.merge(df_1,df_2,left_on='city',right_on='city')
```

```python
df_1.join(df_2,lsuffix='_1', rsuffix='_2') #==> Join here is not correct, you need to do like in the following cell
```

```python
df_1.join(df_2.set_index('city'),lsuffix='_1', rsuffix='_2',on='city') 
# also you can use: how='inner' ==> different values: {‘left’, ‘right’, ‘outer’, ‘inner’}, default ‘left’
```

```python
pd.concat((df_1,df_2)) #by default axis=0
```

```python
#This syntax is not correct:
pd.concat((df_1,df_2),axis=1)
```

```python
# This one is correct
pd.concat((df_1.set_index('city'),df_2.set_index('city')),axis=1)
```
